### Valores Discrepantes (outliers)

```python
lista_idades = [26,30,32,22,26,35,400,20,43,31,23]
```

```python
#Ao fazer a média da lista é evidente que a média é muito discrepantes, comparada
#aos valores do resto da lista, e essa discrepância é por conta do 440 que foi inserido
media = np.mean(lista_idades)
media
```

```python
#Para ver se tem alguma chance de ter um valor discrepante na lista, é preciso
#fazer o desvio padrão. E como foi um valor alto, as chances de ter um valor
#discrepantes são grande
desvio = np.std(lista_idades)
desvio
```

```python
#Calculando o limite superior do outliner
media + 3*desvio
```

```python
#Calculando o limite inferior do outliner
media - 3*desvio
```

```python
#Visualizando de uma forma mais "conservadora" o outliner, utilizando o quartis
#e também utilzando o matplotlib
import matplotlib.pyplot as plt
```

```python
#É evidente que o outliner é aquela bolinha que ultrapassa a linha superior
plt.boxplot(lista_idades)
```

```python
plt.boxplot(df['SALARIO'])
```

```python
 #Para calcular o quartis do Q1 (borda)
 Q1 = df['SALARIO'].quantile(0.25)
 Q1
```

```python
#Para calcular o quartis do Q3 (borda)
Q3 = df['SALARIO'].quantile(0.75)
Q3
```

```python
#Calculando o Intervalo Interquartil (IQR)
IQR = Q3-Q1
IQR
```

```python
#Definindo o limite superior
lim_superior = Q3 + (1.5*IQR)
lim_superior
```

```python
#Definindo o limite Inferior
lim_inferior = Q1 - (1.5*IQR)
lim_inferior
```

```python
#Verificando a faixa salarial, para ver se o limite superior de 22mil se encaixa como
#um outlier, pois no gráfico é mostrado valores grande como outliers, talvez 22mil
#seja um valor muito sucinto
df['FAIXA SALARIAL'].value_counts()
```

```python
#Calculando a média da coluna SALARIO
media_salario = df['SALARIO'].mean()
media_salario
```

```python
#Calculando o desvio padrão da coluna SALARIO
desvio_salario = df['SALARIO'].std()
desvio_salario
```

```python
#Calculando o limite superior utilizando 3 vezes o desvio padrão.
#Essa abordagem considera que os valores acima de 3 desvios padrão da média
#são considerados extremos (outliers) em uma distribuição normal.
#No gráfico, os salários considerados outliers apresentaram valores muito altos.
#Por isso, estamos procurando identificar valores superiores a 22 mil,
#que parecem ser consistentes com o comportamento observado no gráfico
limite_superior = media_salario + (3*desvio_salario)
limite_superior
```

```python
#Conta a quantidade de pessoas por faixa salarial que têm salário acima do limite superior
df[df['SALARIO']>limite_superior]['FAIXA SALARIAL'].value_counts()
```

```python
#Calcula a média dos salários na faixa "de R$ 30.001/mês a R$ 40.000/mês" abaixo do limite superior
media_30_40 = df[(df['FAIXA SALARIAL']=='de R$ 30.001/mês a R$ 40.000/mês')&(df['SALARIO']<limite_superior)]['SALARIO'].mean()
```

```python
#Substitui os salários acima do limite superior por essa média na faixa "de R$ 30.001/mês a R$ 40.000/mês"
df.loc[(df['FAIXA SALARIAL']=='de R$ 30.001/mês a R$ 40.000/mês')&(df['SALARIO']>limite_superior), 'SALARIO'] = media_30_40
```

```python
#Repete a contagem das faixas salariais para verificar alterações após o ajuste
df[df['SALARIO']>limite_superior]['FAIXA SALARIAL'].value_counts()
```

```python
#Calcula a média dos salários na faixa "Acima de R$ 40.001/mês" abaixo do limite superior
media_40 = df[(df['FAIXA SALARIAL']=='Acima de R$ 40.001/mês')&(df['SALARIO']<limite_superior)]['SALARIO'].mean()
```

```python
#Substitui os salários acima do limite superior por essa média na faixa "Acima de R$ 40.001/mês"
df.loc[(df['FAIXA SALARIAL']=='Acima de R$ 40.001/mês')&(df['SALARIO']>limite_superior), 'SALARIO'] = media_40
```

```python
#Conta a quantidade de pessoas por faixa salarial que ainda têm salário acima do limite superior
df[df['SALARIO']>limite_superior]['FAIXA SALARIAL'].value_counts()
```

#Distribuição Amostral e Intervalo de Confiança

```python
#Selecionando a coluna 'SALARIO'
salarios = df['SALARIO']
```

```python
#Exibindo os valores dos salários
salarios
```

```python
#Calculando a média amostral dos salários
media_amostral = np.mean(salarios)
media_amostral
```

```python
#Calculando o desvio padrão amostral dos salários
desvio_amostral = np.std(salarios)
desvio_amostral
```

```python
#Definindo o nível de confiança desejado (95% neste caso)
nivel_de_confianca = 0.95
```

```python
#Obtendo o tamanho da amostra (número de salários)
tamanho_da_amostra = len(salarios)
tamanho_da_amostra
```

```python
#Importando o módulo stats da biblioteca scipy
from scipy import stats
```

```python
#Calculando o erro padrão do intervalo entre as amostras dos nossos dados de salario
erro_padrao = stats.sem(salarios)
erro_padrao
```

```python
#Calculando o intervalo de confiança
#Com 95% de confiança, a média salarial do Brasil está entre 9 mil e 10 mil
intervalo_confianca = stats.t.interval(nivel_de_confianca, tamanho_da_amostra-1, loc=media_amostral, scale=erro_padrao)
intervalo_confianca
```
